```python
import streamlit as st
import pandas as pd

st.title("IoT HVAC â€“ Realtime Dashboard")

# layout
left, right = st.columns([1,2])

with left:
    st.subheader("Latest Sensor Data")
    st.metric("Temperature", "29.8Â°C")
    st.metric("Humidity", "71%")
    st.metric("ML Prediction", "Normal")

with right:
    df = pd.DataFrame({
        "temp": [28,29,29.5,30,29.8],
        "hum": [70,72,73,75,71]
    })
    st.line_chart(df["temp"])
```
